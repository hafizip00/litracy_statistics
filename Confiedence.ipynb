```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pyplot
from scipy import stats
```

```python
data = pd.read_csv("Litracy.csv")
data.head(10)
```

```python
edu_district = data.dropna()
sampled_data = edu_district.sample(n=50,replace=True,random_state=13450)
```

```python
sampled_data
```

```python
sample_mean = sampled_data['Value'].mean()
sample_mean
```

```python
estimated_sample_error = sampled_data['Value'].std() / np.sqrt(sampled_data.shape[0])
```

```python
stats.norm.interval(confidence=0.95, loc=sample_mean, scale=estimated_sample_error)
```

```python
stats.norm.interval(confidence=0.99, loc=sample_mean, scale=estimated_sample_error)
```

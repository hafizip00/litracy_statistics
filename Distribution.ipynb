```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pyplot
```

```python
data = pd.read_csv("Litracy.csv")
data.head(10)
```

```python
import statsmodels.api as sm
from scipy import stats
```

```python
data['Value'].hist()
```

```python
sum_of_all=data['Value'].sum()
```

```python
data["Percent"] = (data['Value']*100)/sum_of_all
```

```python
data["Percent"] = data['Percent']*100
```

```python
data['Percent']
```

```python
mean_data = data['Percent'].mean()
```

```python
data['Value'].std()
```

```python
std_data = data['Percent'].std()
```

```python
upper_limit = mean_data + 1*std_data
lower_limit = mean_data - 1*std_data
((data['Percent'] >= lower_limit) & (data['Percent'] <= upper_limit)).mean()
```

```python
## Data is not Normally Distributed
```

```python
upper_limit = mean_data + 3*std_data
lower_limit = mean_data - 3*std_data
((data['Percent'] >= lower_limit) & (data['Percent'] <= upper_limit)).mean()
```

```python
data['Z-Score'] = stats.zscore(data['Percent'])
```

```python
data[(data['Z-Score'] < -3) | (data['Z-Score'] > 3)]
```
